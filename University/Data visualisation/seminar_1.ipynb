```python
x = np.arange(-2, 10, 0.01)
y = np.arange(-2, 10, 0.01)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z = (np.sin(Y)+np.power(X,2))*np.exp(X+Y)+Y
```

```python
plt.contour (X, Y, Z, colors='green', levels = 1)
plt.colorbar()
```

```python
plt.contourf(X, Y, Z, levels = 1)
plt.colorbar()
```

## Задание 3

```python
from matplotlib import gridspec
```

```python
x1 = np.arange(0,16)
y1 = np.sin(x1)*np.cos(2*x1)
```

```python
x2 = np.arange(-17,5)
y2 = -1*2/x2+np.cos(x2)+np.sin(np.power(x2,3))
```

```python
fig, ax = plt.subplots(nrows=1,ncols = 2, gridspec_kw={'width_ratios': [1, 5]})
ax[0].plot(x1,y1,'r')
plt.grid()
ax[1].plot(x2,y2,'g')
plt.grid()
```

## Задание 4

```python
names = ['Петров', 'Иванов', 'Васечкин', 'Подлипки', 'Раскольников', 'Водянкина']
valuess = [90, 85, 45, 75, 56, 35]
```

```python
N = len(valuess)
theta = np.arange(0,2 * np.pi, 2 * np.pi/N)
plt.axes([0.025, 0.025, 0.95, 0.95], polar=True)
plt.title('Задание 4: построить лепестковую диаграмму')
colors = np.array(['#4bb2c5','#c5b47f','#EAA228','#579575','#839557','#958c12','#953579','#4b5de4'])
bars = plt.bar(theta, valuess, width=(2*np.pi/N), bottom=0.0, color=colors, tick_label = names)
```
