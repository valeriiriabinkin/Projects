```python
def longest_words(file):
    read_file = open(file,'r')
    file_contents = read_file.read()
    a = file_contents.split()
    a = sorted(a, key = lambda x:len(x))
    length = []
    for i in a:
        length.append(len(i))
    count = 0
    max_len = max(length)
    for i in length:
        if i == max_len:
            count+=1
    
    return a[-count]
```

```python
df = pd.DataFrame(columns = ['name', 'id', 'price', 'creation_date'])
b = 0
while True:
    a = get_random_date(start_dt, end_dt)
    c = input("введите цену товара:")
    d = input("введите название мебели:")
    if d.lower() == 'конец' or c.lower() == 'конец':
        conn = sqlite3.connect('new_db.db')
        df.to_sql('furniture', conn, if_exists = 'replace')
        break
    else:
        df.loc[len(df.index)] = [d,b,int(c),a]
        b+=1
```
