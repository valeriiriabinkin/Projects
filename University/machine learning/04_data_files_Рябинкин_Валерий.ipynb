````python
#проверял, как работает поиск username у конкретного пользователя
data[0]['username'] == 'uhebert'
````

````python
def funca(jsonn, a):
    a1 = []
    for i in jsonn:
        if i['username'] == a:
            a1.append(i)
            break
        else:
            raise ValueError('пользователь с заданным username отсутствует')
    return a1
````

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python
from bs4 import BeautifulSoup
````

````python
with open('steps_sample.xml') as f:
    data = BeautifulSoup(f,'xml')
````

````python
data.recipes.find_all('recipe')
````

````python
data.recipes.steps.find_all('step')
````

````python
res = list()
for i in data.recipes.find_all('recipe'):
    ph=[st.next for st in i.steps.find_all('step')]
    res.append({i.find('id').next: ph})
````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
res1 = list()
for i in data.recipes.find_all('recipe'):
    ph = len([st.next for st in i.steps.find_all('step')])
    res1.append({ph: i.find('id').next})
````

````python
key1 = []
for i in res1:
    for key in i.keys():
        key1.append(key)
````

````python
key1
````

````python
val1 = []
for i in res1:
    for value in i.values():
        val1.append(value)
````

````python
val1
````

````python
col_names =  ['key', 'value']
dataa  = pd.DataFrame(columns = col_names)
dataa['key'] = key1
dataa['value'] = val1
````

````python
dataa
````

````python
dataa1 = dataa.groupby(['key'])['value'].apply(', '.join).reset_index()
````

````python
dataa1
````

````python
slovar1 = dict(zip(dataa1['key'], list(dataa1['value'])))
````

````python
slovar1
````
