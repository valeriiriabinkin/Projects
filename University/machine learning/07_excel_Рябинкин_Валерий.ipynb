1. Загрузите данные из файлов `reviews_sample.csv` (__ЛР2__) и `recipes_sample_with_tags_ingredients.csv` (__ЛР5__) в виде `pd.DataFrame`. Обратите внимание на корректное считывание столбца(ов) с индексами. Оставьте в таблице с рецептами следующие столбцы: `id`, `name`, `minutes`, `submitted`, `description`, `n_ingredients`

```python
import pandas as pd
import xlwings as xw
```

```python
rs = pd.read_csv('reviews_sample.csv')
rsw = pd.read_csv('recipes_sample_with_tags_ingredients.csv')
```

```python
rsw = rsw[['id', 'name', 'minutes', 'submitted', 'description', 'n_ingredients']]
```

2. Случайным образом выберите 5% строк из каждой таблицы и сохраните две таблицы на разные листы в один файл `recipes.xlsx`. Дайте листам названия "Рецепты" и "Отзывы", соответствующие содержанию таблиц. 

```python
type(rsw.index)
```

```python
import random
rand1 = random.sample(list(rsw.index), round(len(rsw)*0.05))
rand2 = random.sample(list(rs.index), round(len(rs)*0.05))
```

```python
rand1
```

```python
rsw = rsw.loc[rand1]
rs = rs.loc[rand2]
```

```python
writer = pd.ExcelWriter('recipes.xlsx')
```

```python
type(writer)
```

```python
rsw.to_excel(writer, "Рецепты")
rs.to_excel(writer, "Отзывы")
```

```python
writer.save()
```

3. Используя `xlwings`, добавьте на лист `Рецепты` столбец `seconds_assign`, показывающий время выполнения рецепта в секундах. Выполните задание при помощи присваивания массива значений диапазону ячеек.

```python
time = rsw['minutes']
```

```python
time = [i*60 for i in time]
```

```python
len(time)
```

```python
wb = xw.Book('recipes.xlsx')
```

```python
type(wb)
```

```python
sht = wb.sheets["Рецепты"]
```

```python
type(sht)
```

```python
sht.range('H1').value = 'seconds_assign'
```

```python
sht.range('H2').options(transpose = True).value = time
```

```python
sht.range('H1:H1501').value
```

4. Используя `xlwings`, добавьте на лист `Рецепты` столбец `seconds_formula`, показывающий время выполнения рецепта в секундах. Выполните задание при помощи формул Excel.

```python
sht.range('I1').value = 'seconds_formula'
```

```python
sht.range('I2:I1501').options(transpose = True).formula = f'=(D1:D1501)*60'
```

```python
sht.range('I2:I1501').value
```

5. Добавьте на лист `Рецепты`  столбец `n_reviews`, содержащий кол-во отзывов для этого рецепта. Выполните задание при помощи формул Excel.

```python
sht1 = wb.sheets['Отзывы']
```

```python
sht.range('J1').value = 'n_reviews'
```

```python
r_id = sht1.range('D2').expand('down').address
```

```python
r_id
```

```python
sht.range('J2').formula = f'=COUNTIF(Отзывы!{r_id}, B2)'
```

```python
from xlwings.constants import AutoFillType
xw.Range('J2').api.AutoFill(xw.Range(f"J2:J1501").api,AutoFillType.xlFillDefault)
```

6. Сделайте названия всех добавленных столбцов полужирными и выровняйте по центру ячейки.

```python
sht.range('$A1:$J1').api.HorizontalAlignment = -4108
sht.range('$A1:$J1').api.Font.Bold = True
```

7. Раскрасьте ячейки столбца `minutes` в соответствии со следующим правилом: если рецепт выполняется быстрее 5 минут, то цвет - зеленый; от 5 до 10 минут - жёлтый; и больше 10 - красный.

```python
type(sht.range('D2:D1501')[0])
```

```python
for i in sht.range('D2:D1501'):
    if i.value<5:
        i.color = (0,128,0)
    elif i.value>10:
        i.color = (255,0,0)
    else:
        i.color = (255,255,0)
```
